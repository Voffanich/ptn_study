```python
class CZ():
    
    win_combs = [[0, 1, 2], [0, 3, 6], [0, 4, 8], [1, 4, 7], [2, 5, 8], [2, 4, 6], [3, 4, 5], [6, 7, 8]]
    turn = 1
    
    def __init__ (self, gamer_1, gamer_2):
        self.gamer_x = gamer_1
        self.gamer_o = gamer_2
        
        print("Cells id's:")
        self.field = [str(i) for i in range(1,10)]
        self.draw_field()
        print("LETS BEGIN!")
        
        self.field = [" " for i in range(9)]
        
    def draw_field(self):
        
        for i in range(0,8,3):
            
            print(self.field[i] + " | " + self.field[i+1] + " | " + self.field[i+2])
            if i != 6:
                print("---------")
    
    def check_field(self):
        
        x_set = []
        o_set = []
        x_set_c = []
        o_set_c = []
        fin = 0
        
        for i in range (9):
            if self.field[i] == "X":
                x_set.append(i)
            if self.field[i] == "O":
                o_set.append(i)
         
        print(x_set)
        print(o_set)
       
        for i in self.win_combs:
            if set(x_set).issuperset(set(i)):
                fin = 'X'       
        
        for i in self.win_combs:
            if set(o_set).issuperset(set(i)):
                fin = 'O'  
                
        return fin
    
    def is_filled(self, cell):
        
        if self.field[cell] == 'X' or self.field[cell] == 'O':
            return True
    
    def cell_fill(self):
        
        filled = 0               
                       
        if self.turn == 1:
            
            cell_id = int(input(f'{self.gamer_x}, please, give me the cell id for X:'))-1
            
            while filled == 0:
                
                if cell_id < 0 or cell_id > 8:
                    cell_id = int(input(f'{self.gamer_x}, wrong cell id! Please, give me another cell id 1-9 for X:'))-1
                elif self.is_filled(cell_id):
                    cell_id = int(input(f'{self.gamer_x}, the sell is oquppied! Please, give me another cell id for X:'))-1
                else:
                    self.field[cell_id] = "X"
                    filled = 1     
                                   
            self.turn *= -1

        elif self.turn == -1:

            cell_id = int(input(f'{self.gamer_o}, please, give me the cell id for O:'))-1
            
            while filled == 0:
                
                if cell_id < 0 or cell_id > 8:
                    cell_id = int(input(f'{self.gamer_x}, wrong cell id! Please, give me another cell id 1-9 for O:'))-1
                elif self.is_filled(cell_id):
                    cell_id = int(input(f'{self.gamer_o}, the sell is oquppied! Please, give me another cell id for O:'))-1
                else:
                    self.field[cell_id] = "O"  
                
                    filled = 1                  
            
            self.turn *= -1
    
    def run(self):
        
        fin = 0
        
        for i in range(9):
            
            self.cell_fill()       
            self.draw_field()            
            print('Next turn!')
            print(' ')
            fin = self.check_field()
        
            if fin == 'X':
                print(f'{self.gamer_x} wins! Congratulations!')
                break
            elif fin == 'O':
                print(f'{self.gamer_o} wins! Congratulations!')
                break
            
        if fin == 0:    
            print(f'{self.gamer_o}, {self.gamer_x}, guys, it\'s a draw!')
        
```

```python
game = CZ("Mike", "Nick")
game.run()
```
